```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from matplotlib.ticker import MaxNLocator
```

```python
delta = 0.025
x = np.arange(-5.0, 5.0, delta)
y = np.arange(-5.0, 5.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = X**2 - Y**2
Z2 = 2*X*Y
```

```python
fig, ax = plt.subplots(figsize=(5,5))
CS = ax.contour(X, Y, Z1, colors='b', linestyles='solid', levels=[-30,-25, -20, -15, -10, -5, 0, 5, 10, 15, 20, 25, 30])
ax.contour(X, Y, Z2, colors='r', linestyles='dotted',levels=[-40, -35, -30, -25, -20, -15, -10, -5, 0, 5, 10, 15, 20, 25, 30, 35, 40])
```

```python
xx = np.arange(-40.0, 40.1, 0.1)
yy = np.arange(-40.0, 40.1, 0.1)
XX, YY = np.meshgrid(xx, yy)
Z3 = XX
Z4 = YY
```

```python
fig2, ax2 = plt.subplots(figsize=(5,5))
CS = ax2.contour(XX, YY, Z3, colors='b', linestyles='solid', levels=[-40, -35, -30, -25, -20, -15, -10, -5, 0, 5, 10, 15, 20, 25, 30, 35, 40])
ax2.contour(XX, YY, Z4, colors='r', linestyles='dotted',levels=[-40, -35, -30, -25, -20, -15, -10, -5, 0, 5, 10, 15, 20, 25, 30, 35, 40])
```
