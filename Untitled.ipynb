```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from matplotlib.ticker import MaxNLocator
import cmath
```

```python
x = np.arange(-20, 21, 1)
y = np.arange(-20, 21, 1)

xdense = np.arange(-20, 20.1, 0.1)
ydense = np.arange(-20, 20.1, 0.1)
```

```python
xx, yy = np.meshgrid(x, y)
```

```python
xxd, yyd = np.meshgrid(xdense, ydense)
```

```python
fig, ax = plt.subplots(figsize=(5,5))

for i in np.arange(len(yy)):
    ax.plot(x, yy[i], 'r:')
for j in np.arange(len(xx[1])):
    ax.plot(xx[:,j],y, 'b--')

ax.set_ylim(-10, 10)
ax.set_xlim(-10, 10)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
```

```python
def u(x, y):
    return x*x - y*y

def v(x,y):
    return 2*x*y
```

```python
fig2, ax2 = plt.subplots(figsize=(5,5))

for i in np.arange(len(yyd)//2+1):
    if (i%10 == 0):
        ax2.plot(u(xdense, yyd[i]), v(xdense, yyd[i]), 'r:')
for j in np.arange(len(xxd[0])//2+1):
    if (j%10 == 0):
        ax2.plot(u(xxd[:,j],ydense),v(xxd[:,j],ydense), 'b--')
ax2.set_ylim(-100, 100)
ax2.set_xlim(-100, 100)
ax2.xaxis.set_major_locator(MaxNLocator(integer=True))
ax2.yaxis.set_major_locator(MaxNLocator(integer=True))
```

```python
fig.savefig("grid.pdf", transparent=True,bbox_inches='tight')
```

```python
fig2.savefig("z2.pdf", transparent=True,bbox_inches='tight')
```
