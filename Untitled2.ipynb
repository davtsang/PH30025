```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from matplotlib.ticker import MaxNLocator
import cmath
```

```python
delta = 0.0005
x = np.arange(-1.02, 1.02, delta)
y = np.arange(-1.02, 1.02, delta)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)
zo = 2.0
zoconj = zo.real - 1.0j*zo.imag
Z2 = np.abs((X + 1.0j*Y - zo)/(1 - zoconj*(X+1.0j*Y)))
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.contour(X, Y, Z, colors='r', linestyles='dotted', levels=[0, 0.2, 0.4, 0.6, 0.8])
ax.contour(X, Y, Z, colors='b', linestyles='solid', levels=[0.01,1.0])
ax.set_xlim(-1.02, 1.02)
ax.set_ylim(-1.02, 1.02)
```

```python
fig2, ax2 = plt.subplots(figsize=(5,5))
ax2.contour(X, Y, Z2, colors='r', linestyles='dotted', levels=[0, 0.2, 0.4, 0.6, 0.8])
ax2.contour(X, Y, Z2, colors='b', linestyles='solid', levels=[0.01,1.0])
ax2.set_xlim(-1.02, 1.02)
ax2.set_ylim(-1.02, 1.02)
```
