```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from matplotlib.ticker import MaxNLocator
import cmath
```

```python
u1 = np.arange(-10, 10, 0.001)
v1 = 0.0*u1
w1 = u1 + 1.0j*v1
delta = 0.25
```

```python
uu = np.arange(-18, 18.01, 0.01)
vsteps = np.arange(0+delta, 18+delta, delta)
usteps = np.arange(-18, 18+delta, delta)
vv = np.arange(0, 18.01, 0.01)
stream_list = []
for v in vsteps:
    stream_list.append((uu, np.zeros(len(uu))+v))
phi_list = []
for u in usteps:
    phi_list.append((np.zeros(len(vv))+u, vv))
    
w_stream_list = []
for stream in stream_list:
    w_stream_list.append(stream[0] + 1.0j*stream[1])
w_phi_list=[]
for phi in phi_list:
    w_phi_list.append(phi[0] + 1.0j*phi[1])
```

```python
def ninetydegree(z):
    return cmath.exp(0.5j*np.pi)*2.0*z**0.5
def twoseventydegree(z):
    return cmath.exp(1.5j*np.pi)*z**(1.5)/1.5
```

```python
fig1, ax1 = plt.subplots(figsize=(5,2.5))
ax1.plot(w1.real, w1.imag, 'b-', linewidth=3)
for w_stream in w_stream_list:
    ax1.plot(w_stream.real, w_stream.imag, 'r-')
for w_phi in w_phi_list:
    ax1.plot(w_phi.real, w_phi.imag, 'g:')
ax1.set_xlim(-3, 3)
ax1.set_ylim(-0.1, 2.9)
ax1.axis('off')
```

```python
z_1 = ninetydegree(w1)
z_1_stream_list =[]
z_1_phi_list=[]
for w_stream in w_stream_list:
    z_1_stream_list.append(ninetydegree(w_stream))
for w_phi in w_phi_list:
    z_1_phi_list.append(ninetydegree(w_phi))
```

```python
fig2, ax2 = plt.subplots(figsize=(5,5))
ax2.plot(z_1.real, z_1.imag, 'b-', linewidth=3)
for z_1_stream in z_1_stream_list:
    ax2.plot(z_1_stream.real, z_1_stream.imag, 'r-')
for z_1_phi in z_1_phi_list:
    ax2.plot(z_1_phi.real, z_1_phi.imag, 'g:')
ax2.set_xlim(-3, 0.1)
ax2.set_ylim(-0.1, 3)
ax2.axis('off')
```

```python
z_2 = twoseventydegree(w1)
z_2_stream_list =[]
z_2_phi_list=[]
for w_stream in w_stream_list:
    z_2_stream_list.append(twoseventydegree(w_stream))
for w_phi in w_phi_list:
    z_2_phi_list.append(twoseventydegree(w_phi))
```

```python
fig3, ax3 = plt.subplots(figsize=(5,5))
ax3.plot(z_2.real, z_2.imag, 'b-', linewidth=3)
for z_2_stream in z_2_stream_list:
    ax3.plot(z_2_stream.real, z_2_stream.imag, 'r-')
for z_2_phi in z_2_phi_list:
    ax3.plot(z_2_phi.real, z_2_phi.imag, 'g:')
ax3.set_xlim(-3, 3)
ax3.set_ylim(-3, 3)
ax3.axis('off')
```

```python
fig1.savefig("corner180.pdf", transparent=True,bbox_inches='tight')
fig2.savefig("corner90.pdf", transparent=True,bbox_inches='tight')
fig3.savefig("corner270.pdf", transparent=True,bbox_inches='tight')
```
